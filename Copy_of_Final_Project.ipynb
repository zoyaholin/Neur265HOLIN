```python
#Importing modules
import matplotlib as mpl
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Uploading data
url= 'https://raw.githubusercontent.com/zoyaholin/Neur265HOLIN/main/alzheimers_IDs_final_project.xlsx%20-%20SEA-AD_Cohort_Metadata.csv'
alz_ID= pd.read_csv(url)

url1= 'https://raw.githubusercontent.com/zoyaholin/Neur265HOLIN/main/alzheimers_pathology_final_project.csv'
alz_path= pd.read_csv(url1)

url2= 'https://raw.githubusercontent.com/zoyaholin/Neur265HOLIN/main/Updated%20alz%20-%20Sheet1.csv'
alz_combined= pd.read_csv(url2)
```

```python
#Setting variables
cog_status= alz_ID.iloc[:,17]
sex=alz_ID.iloc[:,4]

AT8positive_area_Layer3 = alz_path.iloc[:,5]
AT8positive_area_Layer4 = alz_path.iloc[:,6]
totalpTDP43_positive_area_Layer1= alz_path.iloc[:,10]
totalpTDP43_positive_area_Layer2= alz_path.iloc[:,11]
```

```python
#Making histogram for AT8 positive layer 3
plt.hist(AT8positive_area_Layer3, color = "pink")
plt.xlabel('AT8 Positive Area Layer 3')
plt.ylabel('Count')
plt.title('Distribution of AT8 Positive Area Layer 3')
```

```python
#Making histogram for AT8 positive layer 4
plt.hist(AT8positive_area_Layer4, color = "pink")
plt.xlabel('AT8 Positive Area Layer 4')
plt.ylabel('Count')
plt.title('Distribution of AT8 Positive Area Layer 4')
```

```python
#Making histogram for total pTDP43_positive_area_Layer1
plt.hist(totalpTDP43_positive_area_Layer1, color = "pink")
plt.xlabel('pTDP43_positive_area_Layer1')
plt.ylabel('Count')
plt.title('Distribution of pTDP43_positive_area_Layer1')
```

```python
#Making histogram for total pTDP43_positive_area_Layer2
plt.hist(totalpTDP43_positive_area_Layer2, color = "pink")
plt.xlabel('pTDP43_positive_area_Layer2')
plt.ylabel('Count')
plt.title('Distribution of pTDP43_positive_area_Layer2')
```

```python
#Making pie chart for cog status grouping variables
colors = ['pink', 'blue']
alz_ID['Cognitive Status'].value_counts().plot(kind='pie', colors=colors)

plt.ylabel('Cognitive Status')
plt.show()
```

```python
#Making pie chart for sex grouping variables
colors = ['pink', 'blue']
alz_ID['Sex'].value_counts().plot(kind='pie', colors=colors)

plt.ylabel('Sex')
plt.show()
```

```python
#scatter of AT8 layer 3/4
alz_combined_length=len(alz_combined)

for i in range(alz_combined_length):
  if alz_combined.iloc[i,2] == 'Dementia':
    plt.scatter(alz_combined.iloc [i,3], alz_combined.iloc [i,4], color = "blue")
  else: 
    plt.scatter(alz_combined.iloc [i,3], alz_combined.iloc [i,4], color = "pink")
    plt.xlabel('total AT8 positive area Layer3 ')
    plt.ylabel('total AT8 positive area Layer4 ')
    plt.title('Total AT8 Positive Area Layers 3 and 4 Plotted Against Cognitive Status')

plt.legend(['Dementia','Non-Dementia'])
```

```python
#Plotting scatter plot for total pTDP43 positive area Layer 1 and total pTDP43 positive area Layer2 
alz_combined_length=len(alz_combined)

for i in range(alz_combined_length):
  if alz_combined.iloc[i,1] == 'Female':
    plt.scatter(alz_combined.iloc [i,5], alz_combined.iloc [i,6], color = "blue")
  else: 
    plt.scatter(alz_combined.iloc [i,5], alz_combined.iloc [i,6], color = "pink")
    plt.xlabel('total pTDP43 positive area Layer 1')
    plt.ylabel('total pTDP43 positive area Layer 2')
    plt.title('Total pTDP43 positive area Layers 1 and 2 Plotted Against Sex')

plt.legend(['Female','Male'])
```

```python

#Making variables into numpy arrays
male_pTDP43_1 = np.array(alz_combined.loc[alz_combined['Sex'] == 'Male', ['total pTDP43 positive area_Layer1']])
female_pTDP43_1 = np.array(alz_combined.loc[alz_combined['Sex'] == 'Female', ['total pTDP43 positive area_Layer1']])

male_pTDP43_2 = np.array(alz_combined.loc[alz_combined['Sex'] == 'Male', ['total pTDP43 positive area_Layer2']])
female_pTDP43_2 = np.array(alz_combined.loc[alz_combined['Sex'] == 'Female', ['total pTDP43 positive area_Layer2']])

Dem_AT8_3 = np.array(alz_combined.loc[alz_combined['Cognitive Status'] == 'Dementia', ['total AT8 positive area_Layer3']])
NoDem_AT8_3 = np.array(alz_combined.loc[alz_combined['Cognitive Status'] == 'No Dementia', ['total AT8 positive area_Layer3']])

Dem_AT8_4 = np.array(alz_combined.loc[alz_combined['Cognitive Status'] == 'Dementia', ['total AT8 positive area_Layer4']])
NoDem_AT8_4 = np.array(alz_combined.loc[alz_combined['Cognitive Status'] == 'No Dementia', ['total AT8 positive area_Layer4']])
```

```python
#Making box plots

fig, ax = plt.subplots(1,4, figsize = (18,4))

#total pTDP43 positive area_Layer1
pTDP43_1 = [female_pTDP43_1.flatten(), male_pTDP43_1.flatten()]
ax[0].boxplot(pTDP43_1)
ax[0].set_xticklabels(['females', 'males'])
ax[0].set(title = 'Total pTDP43 Positive Area Layer1 Dist.')

#total pTDP43 positive area_Layer2
pTDP43_2 = [female_pTDP43_2.flatten(), male_pTDP43_2.flatten()]
ax[1].boxplot(pTDP43_2)
ax[1].set_xticklabels(['females', 'males'])
ax[1].set(title = 'Total pTDP43 Positive Area Layer2 Dist.')

#total AT8 positive area_Layer3
AT8_3 = [Dem_AT8_3.flatten(), NoDem_AT8_3.flatten()]
ax[2].boxplot(AT8_3)
ax[2].set_xticklabels(['nondementia', 'dementia'])
ax[2].set(title = 'Total AT8 positive Area Layer3 Dist.')

#total AT8 positive area_Layer4
AT8_4 = [Dem_AT8_4.flatten(), NoDem_AT8_4.flatten()]
ax[3].boxplot(AT8_4)
ax[3].set_xticklabels(['nondementia', 'dementia'])
ax[3].set(title = 'Total AT8 positive Area Layer4 Dist.')
```

```python
#P-value
print(stats.mannwhitneyu(male_pTDP43_1, female_pTDP43_1))
print(stats.mannwhitneyu(female_pTDP43_2, male_pTDP43_2))
print(stats.mannwhitneyu(NoDem_AT8_3, Dem_AT8_3))
print(stats.mannwhitneyu(NoDem_AT8_4, Dem_AT8_4))
```

```python
#Skew Test 
print(stats.skewtest(AT8positive_area_Layer3))
print(stats.skewtest(AT8positive_area_Layer4))
print(stats.skewtest(totalpTDP43_positive_area_Layer1))
print(stats.skewtest(totalpTDP43_positive_area_Layer2))
```
