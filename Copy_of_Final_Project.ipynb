```python
import matplotlib as mpl
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
url= 'https://raw.githubusercontent.com/zoyaholin/Neur265HOLIN/main/alzheimers_IDs_final_project.xlsx%20-%20SEA-AD_Cohort_Metadata.csv'
alz_ID= pd.read_csv(url)

url1= 'https://raw.githubusercontent.com/zoyaholin/Neur265HOLIN/main/alzheimers_pathology_final_project.csv'
alz_path= pd.read_csv(url1)

url2= 'https://raw.githubusercontent.com/zoyaholin/Neur265HOLIN/main/Updated%20alz%20-%20Sheet1.csv'
alz_combined= pd.read_csv(url2)
```

```python
#Making variables
cog_status= alz_ID.iloc[:,17]
sex=alz_ID.iloc[:,4]

AT8positive_area_Layer3 = alz_path.iloc[:,5]
AT8positive_area_Layer4 = alz_path.iloc[:,6]
totalpTDP43_positive_area_Layer1= alz_path.iloc[:,10]
totalpTDP43_positive_area_Layer2= alz_path.iloc[:,11]
```

```python
#Making histogram for AT8 positive layer 3
plt.hist(AT8positive_area_Layer3, color = "pink")
plt.xlabel('AT8 Positive Area Layer 3')
plt.ylabel('Count')
plt.title('Distribution of AT8 Positive Area Layer 3')
```

```python
#Making histogram for AT8 positive layer 4
plt.hist(AT8positive_area_Layer4, color = "pink")
plt.xlabel('AT8 Positive Area Layer 4')
plt.ylabel('Count')
plt.title('Distribution of AT8 Positive Area Layer 4')
```

```python
#Making histogram for total pTDP43_positive_area_Layer1
plt.hist(totalpTDP43_positive_area_Layer1, color = "pink")
plt.xlabel('pTDP43_positive_area_Layer1')
plt.ylabel('Count')
plt.title('Distribution of pTDP43_positive_area_Layer1')
```

```python
#Making histogram for total pTDP43_positive_area_Layer2
plt.hist(totalpTDP43_positive_area_Layer2, color = "pink")
plt.xlabel('pTDP43_positive_area_Layer2')
plt.ylabel('Count')
plt.title('Distribution of pTDP43_positive_area_Layer2')
```

```python
#Making pie chart for cog status grouping variables
colors = ['pink', 'blue']
alz_ID['Cognitive Status'].value_counts().plot(kind='pie', colors=colors)

plt.ylabel('Cognitive Status')
plt.show()
```

```python
#Making pie chart for sex grouping variables
colors = ['pink', 'blue']
alz_ID['Sex'].value_counts().plot(kind='pie', colors=colors)

plt.ylabel('Sex')
plt.show()
```

```python
#scatter of AT8 layer 3/4
alz_combined_length=len(alz_combined)

for i in range(alz_combined_length):
  if alz_combined.iloc[i,3] == 'Dementia':
    plt.scatter(alz_combined.iloc [i,4], alz_combined.iloc [i,5], color = "blue")
  else: 
    plt.scatter(alz_combined.iloc [i,4], alz_combined.iloc [i,5], color = "pink")
    plt.xlabel('total AT8 positive area Layer3 ')
    plt.ylabel('total AT8 positive area Layer4 ')
    plt.title('Total AT8 Positive Area Layers 3 and 4 Plotted Against Cognitive Status')

plt.legend(['Dementa','Non-Dementia'])
```

```python
#Plotting scatter plot for total AT8 positive area_Layer3 and total AT8 positive area_Layer4
alz_path=len(cog_status)

for i in range(alz_path):
  if cog_status.iloc[i,0] == 'total AT8 positive area_Layer3':
    plt.scatter(cog_status.iloc[i,1], cog_status.iloc[i,2], color = "blue")
  else: 
    plt.scatter(cog_status.iloc[i,1], cog_status.iloc[i,2], color = "red")
    plt.xlabel('Layer 3')
    plt.ylabel('Layer 4')
    plt.title('Relationship between AT8 in Positive Layer 3 and 4')
  plt.show()
```
